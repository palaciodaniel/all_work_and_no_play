```python
import random
```

```python
phrase = "All work and no play makes Jack a dull boy"
```

```python
def display1(phrase):
    for i in range(20):
        print(phrase, end = "\n\n")
```

```python
def display2(phrase):
    for i in range(20):
        print(" " * i * 3 + phrase, end = "\n\n")
```

```python
def display3(phrase):
    for i in range(45):
        print(" " * i, phrase[i: i + 45])
```

```python
def display4(phrase):
    for i in range(40):
        repeat_phrase = (phrase + " ") * 200
        print(" " * (40 - i), repeat_phrase[i:i + 50])
```

```python
def display5(phrase):
    for i in range(55):
        if i % 7 == 0:
            print(" " * 5 + phrase)
        print(phrase)
```

```python
def display6(phrase):
    for i in range(7):
        if i % 3 == 0:
            print((phrase + " ") * 10, end = "\n\n")
        else:
            print((phrase + " ") * 20, end = "\n\n")
```

```python
def display7(phrase):
    for i in range(6):
        if i % 2 == 0:
            print((" " * 20 + phrase + "\n") * 10, end = "\n\n")
        else:
            print((phrase + " ") * 20, end = "\n\n")
```

```python
def display8(phrase):
    for i in range(5):

        cut_phrase = phrase + " "
        print(cut_phrase)
        indent = 0

        for char in cut_phrase:
            if char.isspace() == True:
                indent += 1
                cut_phrase = cut_phrase[cut_phrase.index(char) + 1:]
                print(" " * indent + cut_phrase)
```

```python
display_list = [display1, display2, display3, display4, display5, display6, display7, display8]
```

```python
while display_list:
    random_display = random.choice(display_list)
    random_display(phrase)
    display_list.remove(random_display)
```
